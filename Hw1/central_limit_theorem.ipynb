```python
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
from collections import defaultdict
```

```python
N = 1000
n = [2,5,10]
dice_values = [1,2,3,4,5,6]
```

```python
mu_s = []
mu_values_n2 = []
mu_values_n5 = []
mu_values_n10 = []

for i in range(N):
    
    n_sampled_values_n2 = np.random.choice(dice_values,2)
    n_sampled_values_n5 = np.random.choice(dice_values,5)
    n_sampled_values_n10 = np.random.choice(dice_values,10)
    #mu_values_map[n] = 
    mu_values_n2.append(np.mean(n_sampled_values_n2))
    mu_values_n5.append(np.mean(n_sampled_values_n5))
    mu_values_n10.append(np.mean(n_sampled_values_n10))
    

# print(mu_values_n2)
# print(mu_values_n5)
# print(mu_values_n10)

mu_x_n2 = np.mean(mu_values_n2)
mu_x_n5 = np.mean(mu_values_n5)
mu_x_n10 = np.mean(mu_values_n10)

sigma_x2_n2 = np.power(np.std(mu_values_n2),2)
sigma_x2_n5 = np.power(np.std(mu_values_n5),2)
sigma_x2_n10 = np.power(np.std(mu_values_n10),2)

# print("Mu-x","Sigma-x")
# print(mu_x_n2,mu_x_n5,mu_x_n10)
# print(sigma_x2_n2,sigma_x2_n5,sigma_x2_n10)


# mu = np.mean([mu_x_n2,mu_x_n5,mu_x_n10])
# sigma2 = np.std([sigma_x2_n2,sigma_x2_n5,sigma_x2_n10])  #check this , sigma or sigma^2

# print("Expected Value","Variance")
# print(mu)
# print(sigma2)
```

```python
mean_map = defaultdict(list)

real_mean = np.mean(dice_values)
real_variance = np.var(dice_values)

print("Real mean",real_mean)
print("Real variance",real_variance)

for j in n:
    for i in range(N):

        n_sampled_values = np.random.choice(dice_values,j)
        mean_map[j].append(np.mean(n_sampled_values))
    
    mu_x = np.mean(mean_map[j])
    sigma_x2 = np.var(mean_map[j])
    
    print("For n = %s experimental mean= %s,experimental variance= %s" %(j,mu_x,sigma_x2))
    
    plt.hist(mean_map[j], normed=1)
    xmin,xmax = plt.xlim()
    x = np.linspace(xmin,xmax,100)
    plt.plot(x, stats.norm.pdf(x, mu_x, sigma_x2),'k')
    #plt.ylabel('Sampling frequency')
    plt.show()
    
# print(mu_values_n2)
# print(mu_values_n5)
# print(mu_values_n10)

# mu_x_n2 = np.mean(mu_values_n2)
# mu_x_n5 = np.mean(mu_values_n5)
# mu_x_n10 = np.mean(mu_values_n10)

# sigma_x2_n2 = np.power(np.std(mu_values_n2),2)
# sigma_x2_n5 = np.power(np.std(mu_values_n5),2)
# sigma_x2_n10 = np.power(np.std(mu_values_n10),2)
```
